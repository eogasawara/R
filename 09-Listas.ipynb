```r
#slide 2
weight <- c(60, 72, 57, 90, 95, 72) 
height <- c(1.75, 1.80, 1.65, 1.90, 1.74, 1.91)
subject <- c("A", "B", "C", "D", "E", "F")
```

```r
#slide 3
mybag <- list(weight, height, subject, 0, "a")
mybag
```

```r
#slide 4
n <- length(mybag)
mybag[[n+1]] <- "b"
mybag
```

```r
#slide 5
slice <- mybag[1]
slice
is.list(slice)
```

```r
#slide 6
slice <- mybag[c(1,3)]
slice
is.list(slice)
```

```r
#slide 7
is.vector(slice)
```

```r
#slide 8
h <- mybag[[1]]
h
```

```r
#slide 9
is.vector(h)
is.list(h)
```
