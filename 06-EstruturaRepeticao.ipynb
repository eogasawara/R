```r
#slide 11
compute_bmi <- function(weight, height) {
  i <- 1
  bmi <- 0
  while (i <= length(weight)) {
    bmi[i] <- weight[i]/height[i]^2
    i <- i + 1
  }
  return(bmi)
} 

bmi <- compute_bmi(weight, height)
bmi
```

```r
#slide 12
compute_bmi <- function(weight, height) {
  resposta <- weight/height^2
  return(resposta)
} 

bmi <- compute_bmi(weight, height)
bmi
```
